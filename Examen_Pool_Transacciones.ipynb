## 3. Gestión del Pool de Conexiones (Tareas 1 y 2)
Complete las funciones de inicialización del pool, obtención y devolución de conexiones.


````python
def inicializar_pool():
    """Inicializa el pool de conexiones con la configuración adecuada."""
    global db_pool
    db_pool = psycopg2.pool.SimpleConnectionPool(
        minconn=3,
        maxconn=10,
        **DB_CONFIG
    )

def obtener_conexion():
    try:
        return db_pool.getconn()
    except psycopg2.pool.PoolError:
        print("Error")
        return None

def liberar_conexion(conn):
    """Libera una conexión devolviéndola al pool."""
    db_pool.putconn(conn)
````

## 4. Lógica de Operaciones de Transferencia
### 4.1 Flujo de una Transferencia Bancaria
El flujo de transferencia de la cuenta de Ana (ID 1) a la de Carlos (ID 2) por $500.00 se modela de la siguiente manera:
1. **Verificar saldo:** Asegurar fondos suficientes en la cuenta de origen.
2. **Débito:** Restar el monto de la cuenta de origen.
3. **Crédito:** Sumar el monto en la cuenta de destino.
4. **Historial:** Registrar el estado del proceso.
5. **Auditoría:** Guardar logs de control.

Si ocurre un fallo al intentar acreditar a la cuenta de destino (por ejemplo, si la cuenta destino no existe), se debe usar un `SAVEPOINT` para revertir el débito sin abortar toda la transacción principal.


### 4.2 Verificación de Saldo y Operación de Débito (Código Base)
Ejecute esta celda para registrar las funciones auxiliares del sistema bancario.


````python
def verificar_saldo(conn, cuenta_id, monto):
    """
    Verifica si una cuenta dispone de saldo suficiente para la operación.
    """
    try:
        cursor = conn.cursor()
        cursor.execute(
            "SELECT saldo FROM cuentas WHERE id = %s",
            (cuenta_id,)
        )
        resultado = cursor.fetchone()
        if resultado and resultado[0] >= monto:
            cursor.close()
            return True
        cursor.close()
        return False
    except Exception as e:
        print(f"Error al verificar saldo: {e}")
        return False

def realizar_debito(conn, cuenta_id, monto):
    """
    Aplica la reducción de saldo en la cuenta origen.
    """
    try:
        cursor = conn.cursor()
        cursor.execute(
            "UPDATE cuentas SET saldo = saldo - %s WHERE id = %s AND saldo >= %s",
            (monto, cuenta_id, monto)
        )
        filas_afectadas = cursor.rowcount
        cursor.close()
        return filas_afectadas > 0
    except Exception as e:
        print(f"Error en débito: {e}")
        return False
````

### 4.3 Operación de Crédito con Recuperación (Tarea 3)
Complete la función de acreditación utilizando sentencias de control para puntos de control transaccional.


````python
def realizar_credito_con_savepoint(conn, cuenta_id, monto, transaccion_id):
    """
    Realiza el crédito a la cuenta destino usando un SAVEPOINT para mitigar errores.
    """
    cursor = None
    try:
        cursor = conn.cursor()
        
        # Paso 1: Crear savepoint antes de la operación de crédito
        # TODO: Ejecutar el comando SQL correspondiente para SAVEPOINT antes_credito
        cursor.execute("SAVEPOINT sv_credito;")
        # Paso 2: Intentar el crédito
        
        cursor.execute(
            "UPDATE cuentas SET saldo = saldo + %s WHERE id = %s",
            (monto, cuenta_id)
        )
        
        if cursor.rowcount == 0:
            # Cuenta destino no existe - revertir al punto de control
            # TODO: Ejecutar ROLLBACK TO SAVEPOINT antes_credito
            cursor.execute("ROLLBACK TO SAVEPOINT sv_credito;")
            # Registrar la causa del error en el historial
            cursor.execute(
                "UPDATE transacciones SET estado = 'error_credito', mensaje_error = %s WHERE id = %s",
                (f"Cuenta destino {cuenta_id} no encontrada", transaccion_id)
            )
            return False, "Cuenta destino no encontrada"
        
        # Paso 3: Confirmar liberación del savepoint
        # TODO: Ejecutar RELEASE SAVEPOINT antes_credito
        cursor.execute("RELEASE SAVEPOINT sv_credito;")
        return True, "Crédito exitoso"
        
    except Exception as e:
        # En caso de error inesperado, revertir al punto de control
        # TODO: Ejecutar ROLLBACK TO SAVEPOINT antes_credito
        cursor.execute("ROLLBACK TO SAVEPOINT sv_credito;")
        return False, f"Error en crédito: {str(e)}"
    finally:
        if cursor:
            cursor.close()
````

### 4.4 Orquestador de la Transferencia Bancaria (Tarea 4)
Complete el flujo principal de transferencias coordinando los bloques transaccionales (`BEGIN`, `COMMIT`, `ROLLBACK`).


````python
def realizar_transferencia(conn, origen_id, destino_id, monto):
    """
    Controla el flujo de la transferencia completa y la confirmación final de la transacción.
    """
    resultado = {
        'exito': False,
        'mensaje': '',
        'transaccion_id': None
    }
    
    cursor = None
    try:
        cursor = conn.cursor()
        
        # Paso 1: Iniciar transacción explícita
        # TODO: Ejecutar BEGIN
        cursor.execute("BEGIN;")
        # Paso 2: Verificar saldo
        if not verificar_saldo(conn, origen_id, monto):
            resultado['mensaje'] = 'Saldo insuficiente'
            # TODO: Ejecutar ROLLBACK
            cursor.execute("ROLLBACK;")
            return resultado
        
        # Paso 3: Insertar registro preliminar en transacciones
        cursor.execute(
            """INSERT INTO transacciones 
               (cuenta_origen_id, cuenta_destino_id, monto, tipo_operacion, estado)
               VALUES (%s, %s, %s, 'transferencia', 'en_proceso')
               RETURNING id""",
            (origen_id, destino_id, monto)
        )
        transaccion_id = cursor.fetchone()[0]
        resultado['transaccion_id'] = transaccion_id
        
        # Paso 4: Realizar el débito en origen
        if not realizar_debito(conn, origen_id, monto):
            # TODO: Ejecutar ROLLBACK
            cursor.execute("ROLLBACK;")
            resultado['mensaje'] = 'Error en débito'
            return resultado
        
        # Paso 5: Acreditación en destino (con control de savepoint)
        credito_exitoso, credito_mensaje = realizar_credito_con_savepoint(
            conn, destino_id, monto, transaccion_id
        )
        
        if not credito_exitoso:
            # TODO: Ejecutar ROLLBACK
            cursor.execute("ROLLBACK;")
            resultado['mensaje'] = f'Error en crédito: {credito_mensaje}'
            return resultado
        
        # Paso 6: Actualizar estado de la transacción a completada
        cursor.execute(
            "UPDATE transacciones SET estado = 'completada' WHERE id = %s",
            (transaccion_id,)
        )
        
        # Paso 7: Registrar auditoría
        cursor.execute(
            """INSERT INTO logs_auditoria (accion, descripcion)
               VALUES ('transferencia', %s)""",
            (f"Transferencia de {monto} de cuenta {origen_id} a {destino_id} exitosa",)
        )
        
        # Paso 8: Confirmar toda la transacción
        # TODO: Ejecutar COMMIT
        cursor.execute("COMMIT;")
        resultado['exito'] = True
        resultado['mensaje'] = 'Transferencia completada exitosamente'
        
        return resultado
        
    except Exception as e:
        # Si ocurre un fallo crítico general, deshacer todos los cambios
        # TODO: Ejecutar ROLLBACK
        cursor.execute("ROLLBACK;")
        resultado['mensaje'] = f'Error en transferencia: {str(e)}'
        return resultado
    finally:
        if cursor:
            cursor.close()
````

## 5. Consultas de Monitoreo y Visualización de Datos
Ejecute esta celda para habilitar las funciones auxiliares que imprimen el estado de la base de datos.


````python
def mostrar_saldos(conn):
    """Muestra los saldos actuales de todas las cuentas."""
    cursor = conn.cursor()
    cursor.execute("SELECT id, numero_cuenta, titular, saldo FROM cuentas ORDER BY id")
    for cuenta in cursor.fetchall():
        print(f"Cuenta {cuenta[0]}: {cuenta[1]} - {cuenta[2]} - ${cuenta[3]:.2f}")
    cursor.close()

def mostrar_transacciones(conn):
    """Muestra el historial de transacciones procesadas."""
    cursor = conn.cursor()
    cursor.execute(
        """SELECT t.id, c1.numero_cuenta, c2.numero_cuenta, t.monto, t.estado
           FROM transacciones t
           JOIN cuentas c1 ON t.cuenta_origen_id = c1.id
           JOIN cuentas c2 ON t.cuenta_destino_id = c2.id
           ORDER BY t.id DESC LIMIT 5"""
    )
    for trans in cursor.fetchall():
        print(f"Transacción {trans[0]}: {trans[1]} -> {trans[2]} | ${trans[3]:.2f} | {trans[4]}")
    cursor.close()

def mostrar_logs(conn):
    """Muestra las últimas entradas en los logs de auditoría."""
    cursor = conn.cursor()
    cursor.execute("SELECT id, accion, descripcion, fecha FROM logs_auditoria ORDER BY id DESC LIMIT 5")
    for log in cursor.fetchall():
        print(f"Log {log[0]}: {log[1]} - {log[2]} ({log[3]})")
    cursor.close()
````
